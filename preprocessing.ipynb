```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('shapeOfYouComments.csv')  #csv to clean
df.tail()
```

```python
df3 = df[['Author','Comment']].copy()
df3.head()
```

```python
#remove html tags in comments column
df3['Comment'] = df3['Comment'].str.replace(r'<[^<>]*>', ' ', regex=True)  
```

```python
df3.tail(10)
```

```python
#make a dictionary with emoji as key and its text description as value of all emojis in the comments and author columns
import demoji
text =" "
emojiDic ={}
for i in range(len(df3['Comment'])):
    text = text + str(df3['Comment'][i])
for i in range(len(df3['Author'])):
    text = text + str(df3['Author'][i])
emojiDic = demoji.findall(text)
```

```python
print(emojiDic)
```

```python
df3.replace(emojiDic,inplace=True,regex=True)
```

```python
df3.tail(50) #succesfully converted all emojis to text
```

```python
#replace links from comments with blank space
df3['Comment']= df3['Comment'].str.replace('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ',regex=True)
```

```python
df3.tail(50) #observing dataframe
```

```python
#remove non english alphabets
df3 = df3.replace('[^a-zA-Z0-9.,:\d\s]',' ',regex=True)
```

```python
#replacing empty with 0
df3 = df3.replace(r'^\s+$',0, regex=True)
```

```python
df3.tail(50) #observing dataframe
```

```python
df3.to_csv('cleaned_data.csv',index=False)
```
